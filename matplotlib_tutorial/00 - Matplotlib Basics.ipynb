## Matplotlib

**Matplotlib** 
- is a Python package for 2D plotting that generates graphs.
- was modeled on MATLAB.

#### Merits 
1. open source, so no license to pay
2. very customizable and extensible
3. integrated with LaTeX markup
4. cross-platform and portable

#### Output formats
* Raster Images - PNG, JPG, BMP
* Vector Images - _resolution independent_, typesetting, CAD

#### Backends
- the part of Matplotlib that works behind the scenes
- consists of two parts:
  * **Renderer**: does the drawing
    - Standard: Anti-Grain Geometry (AGG) - responsible for the appearance of Matplotlib graphs
  * **Canvas**: destination of the figure
    - Provided with the GUI libraries and any of them can use the AGG Renderer along with their other rendering engines
    * GTKAgg, GTK, GTKCairo, WxAgg, WX, TkAgg, QtAgg, Qt4Agg, FLTKAgg

```python
# -- Write all the imports in this cell
%matplotlib inline
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
import mpld3

mpld3.enable_notebook()
```

```python
plt.plot([1, 3, 2, 4])

plt.show()
```

We only specified a list of values that represent the vertical coordinates of the points to be plotted. 

Matplotlib will use an implicit horizontal values list: [0, N-1]

```python
# We are modeling a parabola y = x^2
x = range(6)

plt.plot(x, [xi ** 2 for xi in x])
plt.show()
```

Aside from the vertical coordinates of the points, we also specified the horizontal coordinates.

The coordinate (x, y) corresponds to the respective indices of the two lists.

But as what we can see, the graph is quite "edgy". So if we want to have a smoother one:

```python
x = np.arange(0.0, 6.0, 0.01)

plt.plot(x, [xi ** 2 for xi in x])
```

We used a numpy function _np.arange_ which returns evenly spaced values within a given interval.

We did this so that we could have a wider range of values and therefore, a smoother graph since we have more values to plot.

```python
# Multiline plots

x = range(1, 5)

plt.plot(x, [xi * 1.5 for xi in x])
plt.plot(x, [xi * 3.0 for xi in x])
plt.plot(x, [xi / 3.0 for xi in x])
plt.plot(x, [xi for xi in x], color="forestgreen")

plt.show()
```

We can also plot multiple lines in a single graph. Matplotlib automatically chooses the color for each plot but there is an option to change the color by using _color_ argument.

It turns out that we can also plot multiple lines just by using a single _plot_ call as specified by the call above _(x, y, x, y, ...)_

```python
x = np.arange(1,5)

plt.grid(True)

plt.plot(x, x * 1.5, x, x * 3.0, x, x / 3.0)
```

In here, we set the plt.grid to True, which will render the grid.

Sometimes, list comprehensions can be kind of messy. We can use _np.arange()_ function that we can use to perform list operations.

```python
x = np.arange(1,5)

plt.plot(x, x * 1.5, x, x * 3.0, x, x / 3.0)

print("Bounds of our axis (xmin, xmax, ymin, ymax): ", plt.axis())

plt.axis([0, 5, -1, 13])
```

In this code snippet, we just changed the limits of our axes which resulted to the graph above.

```python
x = np.arange(1,5)

plt.plot(x, x * 1.5, x, x * 3.0, x, x / 3.0)

print("Bounds of our axis (xmin, xmax, ymin, ymax): ", plt.axis())

plt.axis(xmin=1)
```

If you want to change just one limit, instead of passing an array of four values, we can change it by just specifying its keyword argument.

```python
x = np.arange(1,5)

plt.plot(x, x * 1.5, x, x * 3.0, x, x / 3.0)

print("Bounds of our axis (xmin, xmax, ymin, ymax): ", plt.axis())

plt.xlim(xmin=1)
```

Or if we just want to change the limit/s of a single axis, we can use _plt.xlim()_ or _plt.ylim()_.

```python
x = np.arange(1,5)

plt.plot(x, [xi * 1.5 for xi in x], label="Normal")
plt.plot(x, [xi * 3.0 for xi in x], label="Fast")
plt.plot(x, [xi / 3.0 for xi in x], label="Slow")

print("Bounds of our axis (xmin, xmax, ymin, ymax): ", plt.axis())

plt.xlim(xmin=1)

plt.xlabel("This is the x-axis")
plt.ylabel("This is the y-axis")

plt.title("Multiline plots")
plt.legend()
```

We added a keyword argument, _label_ in _plt.plot()_ function so that we can add it to the legend which is shown using the _plt.legend()_ call. Alternatively, if we do not want to add label for each plots. We can put an array argument that contains the labels for each plot. **Note** that it the labels must be specified in the same order as the created plots. We can do plt.legend(['Normal', 'Fast', 'Slow]).

We also added labels to our x-axis and y-axis by calling the functions _plt.xlabel()_ and _plt.ylabel()_

Lastly, we added title by using _plt.title()_

```python
x = np.arange(1,5)

plt.plot(x, [xi * 1.5 for xi in x], label="Normal")
plt.plot(x, [xi * 3.0 for xi in x], label="Fast")
plt.plot(x, [xi / 3.0 for xi in x], label="Slow")

print("Bounds of our axis (xmin, xmax, ymin, ymax): ", plt.axis())

plt.xlim(xmin=1)

plt.xlabel("This is the x-axis")
plt.ylabel("This is the y-axis")

plt.title("Multiline plots")
```

This is just the same code as above except that we specified a keyword argument, _loc_ to the _plt.legend()_ call. There are several positions where we can place our legend:

* best
* upper right
* upper left
* lower left
* lower right
* right
* center left
* center right
* lower center
* upper center
* center

We can also supply the _loc_ argument with a tuple that contains the coordinates where we want to place our legend. **Note** that the coordinates should be between 0 and 1 only. But we can use some offset like -.1 or so.

```python
x = np.arange(1, 5)

plt.plot(x, x * 1.5, label='Normal')
plt.plot(x, x * 3.0, label='Fast')
plt.plot(x, x / 3.0, label='Slow')

plt.grid(True)

plt.xlim(xmin=1)

plt.legend()

plt.ylabel('Values Measured')
plt.xlabel('Samples')

plt.title('Sample Growth of a Measure')

plt.savefig("../outputs/00-Matplotlib-Basics_1.png", dpi=200, transparent=True)
```
